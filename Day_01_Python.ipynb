Operators in Python

```python
##1. Arthmetic Operators

a = 10
b = 3

print(a + b)   # Addition → 13
print(a - b)   # Subtraction → 7
print(a * b)   # Multiplication → 30
print(a / b)   # Division → 3.33
print(a % b)   # Modulus → 1
print(a ** b)  # Power → 1000
print(a // b)  # Floor Division → 3
```

```python
##Comparison Operators
x = 5
y = 10

print(x == y)   # False
print(x != y)   # True
print(x > y)    # False
print(x < y)    # True
print(x >= 5)   # True
print(y <= 8)   # False
```

```python
##Logical Operators
age = 20
has_id = True

print(age > 18 and has_id)  # True
print(age < 18 or has_id)   # True
print(not has_id)           # False
```

```python
##Assignment Operators
num = 10

num += 5   # num = num + 5
print(num) # 15

num *= 2
print(num) # 30

num -= 10
print(num) # 20

num /= 2
print(num) # 10.0
```

Control Flow


```python
##if / elif / else
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
##for Loop
for i in range(5):
    print("Number:", i)

##Loop through a list:
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)

```

```python
##while Loop
count = 1

while count <= 5:
    print("Count is:", count)
    count += 1
```

break, continue, pass

```python
##break → stop loop
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
##continue → skip current iteration
for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
##pass → placeholder
for i in range(3):
    pass  # Will add code later
```

Data Structures

```python
##Lists
numbers = [1, 2, 3, 4]
numbers.append(5)
numbers.remove(2)

print(numbers)      # [1, 3, 4, 5]
print(numbers[0])   # 1
```

```python
##Tuples (Immutable)
point = (10, 20)

print(point[0])  # 10
# point[0] = 5 Error (cannot change)
```

```python
##Sets (Unique values)
nums = {1, 2, 3, 3, 4}
nums.add(5)
nums.remove(2)

print(nums)  # {1, 3, 4, 5}
```

```python
##Dictionaries
student = {
    "name": "Rahul",
    "age": 21,
    "course": "ML"
}

print(student["name"])  # Rahul

student["age"] = 22
student["grade"] = "A"

print(student)
```

```python
##List Comprehensions
# Square of numbers 0–5
squares = [x**2 for x in range(6)]
print(squares)

# Even numbers only
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
##Defining a Basic Function
def greet():
    print("Hello, welcome to Machine Learning!")

greet()  # Calling the function
```

```python
####Function with Parameters
def greet_user(name):
    print("Hello", name)

greet_user("Parul")
greet_user("Rahul")
```

```python
##Function with Return Value
def add(a, b):
    return a + b

result = add(5, 3)
print("Sum is:", result)
```

```python
##Default Arguments
def power(base, exponent=2):
    return base ** exponent

print(power(5))      # 5^2 = 25
print(power(5, 3))   # 5^3 = 125
```

```python
##*args (Multiple Positional Arguments)
def total_sum(*numbers):
    return sum(numbers)

print(total_sum(1, 2, 3))        # 6
print(total_sum(5, 10, 15, 20))  # 50
```

```python
##*kwargs (Multiple Keyword Arguments)
def student_info(**details):
    for key, value in details.items():
        print(key, ":", value)

student_info(name="Parul", age=21, course="ML")
```

```python
##Lambda Functions
square = lambda x: x * x
print(square(6))  # 36

##With two inputs:
add = lambda a, b: a + b
print(add(4, 7))  # 11
```

```python
##Recursion (Function Calling Itself)
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # 120
```
